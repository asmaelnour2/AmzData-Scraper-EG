```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
fathi=webdriver.Chrome()
fathi.get("https://www.amazon.eg/s?i=electronics&rh=n%3A21832883031%2Cp_123%3A46655&pf_rd_i=21832883031&pf_rd_m=A1ZVRGNO5AYLOV&pf_rd_p=deca76a5-c5fe-453b-ac5f-5b72e8f03038&pf_rd_r=ZHC7WMCHZSFS0QGQ36XX&pf_rd_s=merchandised-search-9&ref=wirelessbrands_EG_1%22")
```

```python
items=fathi.find_elements(By.XPATH,"//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']/span")
prices=fathi.find_elements(By.XPATH,"//span[@class='a-price-whole']")
```

```python
for item in items:
    print(item.text)
for price in prices:
    print(price.text)  
 
print(len(items))
print(len(prices))
```

```python
products = fathi.find_elements(By.XPATH,"//div[@data-component-type='s-search-result']")

for product in products:

    try:
        name = product.find_element(By.XPATH,".//h2/span").text
    except:
        name = "No name"

    try:
        price = product.find_element(By.XPATH,".//span[@class='a-price-whole']").text
    except:
        price = "No price"

    try:
        rating = product.find_element(By.XPATH,".//span[@class='a-icon-alt']").text
    except:
        rating = "No rating"
    
    try:
        reviews = product.find_element(By.XPATH,".//span[@class='a-size-base s-underline-text']").text
    except:
        reviews = "No reviews"

    try:
        image = product.find_element(By.XPATH,".//img").get_attribute("src")
    except:
        image = "No image"

    

    print("Name:", name)
    print("Price:", price)
    print("Rating:", rating)
    print("Reviews:", reviews)
    print("Image:", image)
```
